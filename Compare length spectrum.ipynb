```python
import snappy
import csv
import time
```

```python
groups=[]
with open('groups_of_same_volume_alex_knotFloer_coversdeg7_subgrpIdx6.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if row[0]!='group':
            groups.append(eval(row[0]))

len(groups)
```

```python
def compare_length_spectra(M1,M2,cutoff=2.0,try_hard=False):
    '''
    Takes as input two manifolds and the optional cutoff value.
    Returns True if the length spectra agree up to the cutoff and otherwise False.
    '''
    if try_hard==False:
        spec_M1=M1.length_spectrum(cutoff=cutoff)
        spec_M2=M2.length_spectrum(cutoff=cutoff)
        if len(spec_M1)!=len(spec_M2):
            return False
        for i in range(len(spec_M1)):
            if spec_M[i].length.numerical_approx(digits=5)!=spec_N[i].length.numerical_approx(digits=5):
                return False
        return True

    if try_hard:
        M=M1.high_precision()
        N=M2.high_precision()
        spec_M=M.length_spectrum(cutoff=cutoff)
        spec_N=N.length_spectrum(cutoff=cutoff)
        if len(spec_M)!=len(spec_N):
            return False
        for i in range(len(spec_M)):
            if spec_M[i].length.numerical_approx(digits=10)!=spec_N[i].length.numerical_approx(digits=10):
                return False
        return True
```

```python
total_time=time.time()
max_length=3
different_length_spectra=[]
print('We are checking the length spectra up to length:',max_length)
for [knot1,knot2] in groups:
    start=time.time()
    K1=snappy.Manifold(knot1)
    K2=snappy.Manifold(knot2)
    K1.dehn_fill((0,1))
    K2.dehn_fill((0,1))
    try:
        spec=compare_length_spectra(K1,K2,cutoff=max_length,try_hard=False)
        print('We compare the knots:',knot1,knot2)
        if spec:
            print('The length spectra up to length ',max_length,' are the same.')
        if spec==False:
            print('The length spectra up to length ',max_length,' are different.')
            different_length_spectra.append([[knot1,knot2],'The length spectra up to length '+str(max_length)+' are different.'])
        print('Time taken:',(time.time()-start)/60,' minutes')
    except:
        pass
        #print('An error occured.')
    
print('We could distinguish ',len(different_length_spectra),' pairs.')
print('Total time taken:',(time.time()-total_time)/60,' minutes')
```
